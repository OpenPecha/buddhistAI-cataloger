```python
import json

# Path to the text file containing volume IDs
txt_file_path = r'd:\work\buddhistAI tools\cataloger\volume_ids_back.txt'

# Read the content of the text file
with open(txt_file_path, 'r', encoding='utf-8') as file:
    # Assuming each line is a volume ID, strip newline characters
    volume_ids = [line.strip() for line in file if line.strip()]

# Convert the list of volume IDs to JSON
json_content = json.dumps(volume_ids, ensure_ascii=False, indent=2)

# Print or save the JSON as needed
print(json_content)

# Optionally, save the JSON to a file
# with open('volume_ids_back.json', 'w', encoding='utf-8') as json_file:
#     json_file.write(json_content)
```

```python
import json
from pathlib import Path

txt_path = Path("seeders/volume_ids_back.txt")
json_path = Path("seeders/volume_ids_back.json")

volume_ids = [
    line.strip()
    for line in txt_path.read_text(encoding="utf-8").splitlines()
    if line.strip()
]

volume_ids_json = json.dumps(volume_ids, indent=2, ensure_ascii=False)
json_path.write_text(volume_ids_json, encoding="utf-8")

print(f"Read {len(volume_ids)} volume IDs from {txt_path}")
print(f"Wrote {json_path}")
print(volume_ids_json[:500] + ("..." if len(volume_ids_json) > 500 else ""))
```
